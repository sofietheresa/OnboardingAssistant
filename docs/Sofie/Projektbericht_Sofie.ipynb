## Abstract 

Dieser Projektbericht beschreibt die Erfahrungen bei der Entwicklung und Implementierung einer cloud-nativen Anwendung mit IBM WatsonX-Technologien. Das Projekt "Boardy" ist ein intelligenter Onboarding-Assistent für neue Mitarbeiter, der moderne KI-Technologien wie Retrieval-Augmented Generation (RAG) und semantische Suche nutzt. Die Anwendung wurde vollständig in der IBM Cloud deployed und demonstriert die praktische Anwendung von Large Language Models (LLMs) in einem produktiven Umfeld.


## 1. Einleitung

### 1.1 Motivation 

Die Einarbeitung neuer Mitarbeiter:innen stellt für Unternehmen einen kritischen Erfolgsfaktor dar. Ein strukturierter Onboarding-Prozess trägt nicht nur dazu bei, dass sich neue Teammitglieder schneller zurechtfinden, sondern steigert auch langfristig ihre Produktivität, Zufriedenheit und Bindung an das Unternehmen. In der Praxis ist das Onboarding jedoch häufig fragmentiert: Informationen liegen verteilt in unterschiedlichen Dokumenten, HR-Systemen oder werden mündlich vermittelt. Dies führt leicht zu Verzögerungen, Unsicherheiten und Mehraufwand – sowohl auf Seiten der neuen Mitarbeitenden als auch der betreuenden Führungskräfte.

Vor diesem Hintergrund gewinnt der Einsatz von KI-gestützten Lösungen zunehmend an Bedeutung. Systeme wie IBM WatsonX ermöglichen es, Unternehmenswissen strukturiert zugänglich zu machen und über dialogorientierte Schnittstellen bereitzustellen. Chatbots, die auf firmenspezifische Dokumente zugreifen können, bieten dabei eine besonders intuitive Möglichkeit, Informationen bedarfsgerecht und kontextsensitiv abzurufen.

Das Projekt **Boardy** verfolgt das Ziel, diesen Ansatz für den Onboarding-Prozess nutzbar zu machen. Durch die Integration von WatsonX-Technologien soll ein einfacher, interaktiver Zugang zu relevanten Unternehmensinformationen geschaffen werden. Damit wird nicht nur der Wissenstransfer beschleunigt, sondern auch eine zeitgemäße, digitale Erfahrung im Sinne der Employee Experience unterstützt.

### 1.2 Projektübersicht 

Die im Rahmen des Moduls "AI Labs" entwickelte Anwendung trägt den Namen **Boardy** und versteht sich als KI-gestützte Onboarding-Lösung. Kernidee ist es, neuen Mitarbeiter:innen einen direkten Zugang zu zentralen Informationen des Unternehmens zu bieten, indem sie mit bereitgestellten Dokumenten in Form eines Chats interagieren können. Dabei sollen sowohl informelle als auch offizielle Quellen genutzt werden, um dem Mitarbeiter alle relevanten Informationen zur Verfügung zu stellen. 

Diese Arbeit soll einen Einblick in die Erarbeitung der Anwendung geben.

# 2. Konzept 

Das Onboarding-Assistenzsystem *Boardy* wurde auf Grundlage der  vorausgehenden systematischen Bedarfsanalyse entwickelt. Ziel ist es, neuen Mitarbeiter\:innen eine niedrigschwellige, zentrale und zugleich leistungsfähige Unterstützung während der Einarbeitung anzubieten. Dabei steht die Entlastung von Führungskräften ebenso im Fokus wie die Förderung einer positiven, integrierenden Onboarding-Erfahrung.

## 2.1 Leitidee

Im Kern versteht sich *Boardy* als dialogorientierter **digitaler Begleiter**, der in Form einer Chat-Oberfläche zur Verfügung steht. Die Gestaltung orientiert sich an gängigen Messenger-Diensten und KI-basierten Chatbots, um eine vertraute, barrierearme Nutzung zu ermöglichen. Die Kombination aus natürlicher Spracheingabe, Sprachsteuerung und strukturierter Wissensbasis schafft ein Werkzeug, das sowohl funktionale Fragen klärt als auch soziale und emotionale Bedürfnisse neuer Mitarbeitender adressiert.

## 2.2 Funktionen und Nutzen 

Die Funktionen von *Boardy* sind direkt aus den ermittelten Anforderungen der Zielgruppe abgeleitet:

* **Dialogbasierte Interaktion**: Fragen können flexibel per Texteingabe oder Sprache gestellt werden.
* **Mehrere parallele Chats**: Nutzer\:innen können Themen voneinander trennen und Informationen gezielt wiederfinden.
* **Erweiterbare Wissensbasis**: Offizielle Dokumente, Lagepläne und informelles Wissen von Kolleg\:innen bilden die Grundlage. Neue Inhalte lassen sich direkt über den Chat ergänzen.
* **Proaktive Unterstützung**: *Boardy* begrüßt neue Mitarbeitende aktiv, erinnert an wichtige Schritte und vermittelt das Gefühl, begleitet zu werden.
* **Eskalation und Weiterleitung**: Offene oder komplexe Fragen können an HR oder andere zuständige Stellen weitergeleitet werden.
* **Quellenangaben und Transparenz**: Antworten werden möglichst mit Belegen versehen, um Nachvollziehbarkeit und Vertrauen zu schaffen.
* **Alltagsorientierte Hilfen**: Informationen zu Arbeitsprozessen, Lagepläne, Freizeitangebote oder Kantinenempfehlungen erleichtern die Orientierung.
* **Feedback-Kanal**: Mitarbeitende können ihre Eindrücke – anonym oder offen – zurückmelden, sodass das Unternehmen gezielt nachsteuern kann.

## 2.3 Verknüpfung von Nutzer- und Unternehmenszielen 

Die Anforderungen neuer Mitarbeitender lassen sich in drei Dimensionen zusammenfassen:

* **funktional** (z. B. Prozesse verstehen, Räume finden, Tools nutzen),
* **sozial** (z. B. Kolleg\:innen kennenlernen, Unternehmenskultur verstehen),
* **emotional** (z. B. Sicherheit empfinden, sich wertgeschätzt fühlen).

Aus Unternehmenssicht stehen vor allem Effizienz, Kostenreduktion und eine schnelle Integration der neuen Kräfte im Vordergrund. *Boardy* adressiert diese beiden Perspektiven gleichzeitig, indem er Orientierung bietet, Informationslücken schließt und so die Einarbeitung beschleunigt.

## 2.4 Persönlichkeit und Gestaltung 

*Boardy* vermittelt durch sein **freundliches, kompetentes und empathisches Auftreten** Verlässlichkeit und Nähe. Die visuelle Gestaltung setzt auf vertrauensfördernde Farben, klare Kontraste und eine an bekannte Messenger angelehnte Bedienlogik. Damit wirkt das System zugänglich, modern und leicht erlernbar.

Durch diese Synthese entsteht ein Assistent, der nicht nur Informationsquelle, sondern auch sozialer und emotionaler Begleiter im Onboarding ist – und so den Übergang in die neue Arbeitsumgebung spürbar erleichtert.



# 3. Technologisches Konzept 

Das zuvor beschriebene allgemeine Konzept von *Boardy* zielt wie zuvor beschrieben darauf ab, neuen Mitarbeiter:innen während des Onboardings einen **intuitiven, dialogbasierten Zugang** zu allen relevanten Informationen zu bieten. Im Vordergrund stehen eine vertraute **Chat-Interaktion**, die **Erweiterbarkeit der Wissensdatenbank** sowie eine **proaktive Begleitung** durch den Chatbot. Diese konzeptionellen Ideen werden im technologischen Konzept durch moderne Methoden der **Retrieval-Augmented Generation (RAG)**, semantische Suche in **Vektor-Datenbanken** sowie die Integration von **IBM WatsonX.ai** konkretisiert.

## 3.1 Verknüpfung von Konzept und Technologie 

Die wesentlichen Anforderungen aus dem allgemeinen Konzept lassen sich wie folgt auf die technische Umsetzung abbilden:

| Allgemeines Konzept (Funktion)                              | Technologische Realisierung                                                                   | Literaturbezug                                                     |
| ----------------------------------------------------------- | --------------------------------------------------------------------------------------------- | ------------------------------------------------------------------ |
| **Chat-Interface** mit Text- und Spracheingabe              | React-Frontend mit Speech-to-Text-Integration und REST-API-Anbindung an FastAPI               | –                                                                  |
| **Mehrere parallele Chats**                                 | Session-Management im Backend (FastAPI)                                                       | –                                                                  |
| **Wissensdatenbank** mit formalen und informellen Inhalten  | PostgreSQL mit pgvector für semantische Suche                                                 | vgl. Chowdhury (2010)                                              |
| **Erweiterbarkeit über Chat**                               | Upload- & Annotation-Funktionen im Backend, Integration neuer Dokumente in Embedding-Workflow | vgl. Lewis et al. (2020)                                           |
| **Proaktive Unterstützung** (z. B. Begrüßung am ersten Tag) | Event-gesteuerte Nachrichten aus dem Backend                                                  | –                                                                  |

Damit wird deutlich, dass das **funktionale Konzept** nicht isoliert steht, sondern durch aktuelle KI- und Cloud-Technologien gestützt wird.

## 3.2 RAG als technologische Basis 

Ein zentrales technisches Fundament bildet der Ansatz der **Retrieval-Augmented Generation (RAG)**. Dieser kombiniert klassische **Information-Retrieval-Methoden** mit modernen **Large Language Models (LLMs)**. Während reine LLMs oft unter **Halluzinationen** leiden, ermöglicht RAG eine verlässliche Integration von **unternehmensspezifischem Wissen** in die Antwortgenerierung. Damit wird gewährleistet, dass neue Mitarbeitende nicht nur plausible, sondern auch korrekte und belegbare Informationen erhalten.

## 3.3 Wissensdatenbank und semantische Suche 

Die **Wissensdatenbank** stellt das Herzstück des Systems dar. Im Unterschied zu klassischen FAQ-Bots, die auf vordefinierten Regeln beruhen, nutzt *Boardy* eine **semantische Suche auf Vektor-Ebene**. Mithilfe von Embeddings werden Dokumenteninhalte in hochdimensionale Vektoren transformiert, sodass semantisch ähnliche Anfragen auch dann erkannt werden, wenn unterschiedliche Formulierungen verwendet werden (vgl. Reimers & Gurevych, 2019). Für die Implementierung kommt die **pgvector-Erweiterung** für PostgreSQL zum Einsatz, die eine effiziente Ähnlichkeitssuche ermöglicht.

## 3.4 Cloud-native Architektur 

Um eine skalierbare und robuste Lösung zu gewährleisten, folgt *Boardy* einer **cloud-nativen Architektur**. Prinzipien wie **Containerisierung**, **Microservices** und **API-First-Entwicklung** (vgl. Hohpe & Woolf, 2003; Fielding, 2000) stellen sicher, dass die Anwendung flexibel deployt und in Unternehmensinfrastrukturen integriert werden kann. Die Nutzung von **IBM WatsonX.ai** bietet zusätzlich Enterprise-Sicherheitsstandards sowie Zugriff auf optimierte **Granite-Modelle** für Textgenerierung und Embeddings.

## 3.5 Nutzererlebnis durch technologische Mittel 

Die technologische Umsetzung ist eng mit der konzeptionellen Leitidee verknüpft:

* Die **Proaktivität des Chatbots** wird durch Event-getriebene Nachrichten umgesetzt.
* Die **Vertrauenswürdigkeit der Antworten** wird durch Quellenangaben innerhalb der RAG-Pipeline gestützt.
* Die **Flexibilität der Wissensbasis** erlaubt es, sowohl offizielle Dokumente als auch informelles Wissen direkt im Chat einzupflegen.

Damit schließt sich der Kreis zwischen **konzeptionellem Anspruch** und **technischer Realisierung**: *Boardy* ist nicht nur ein Chatbot, sondern ein **digitaler Mitarbeiterberater**, der durch KI-Methoden wie RAG und semantische Suche in eine moderne cloud-native Architektur eingebettet wird.




## 4. Implementierung 
Die Implementierung der Anwendung erfolgte entlang zweier Hauptstränge: **Frontend** und **Backend**. Beide Komponenten wurden weitgehend parallel entwickelt, wobei das Frontend zunächst unabhängig vom Backend prototypisch umgesetzt wurde. Auf diese Weise konnte frühzeitig eine benutzerfreundliche Oberfläche entworfen und getestet werden.

### 4.1 Frontend-Implementierung 

Die Entwicklung des Frontends orientierte sich an den Prinzipien einer nutzerzentrierten Gestaltung. In Anlehnung an das Modell von **ProRiA und Pichl (2019)** wurde ein Vorgehen gewählt, das neben funktionalen Anforderungen auch ästhetische und emotionale Dimensionen berücksichtigt. Das Modell hebt hervor, dass Benutzeroberflächen nicht nur effizient und funktional, sondern zugleich ansprechend, verständlich und vertrauensfördernd gestaltet werden müssen. Diese Leitidee prägte den gesamten Gestaltungs- und Umsetzungsprozess.

Ziel war es, eine vertraute Chat-Oberfläche zu entwickeln, die sich an etablierten Messenger-Diensten orientiert und dadurch eine niedrige Einstiegshürde für Nutzer schafft.

In der Literatur wird darauf hingewiesen, dass generierter Code häufig weniger strukturiert ist und eine erhöhte **technische Schuld** verursachen kann (vgl. Senn, 2019). ## 4.2 Backend-Implementierung

### 4.2.1 Technisches Konzept der Backend-Architektur 

Das Backend von *Boardy* basiert auf einer **microservice-orientierten Architektur**, die für die Anforderungen eines KI-gestützten Chat-Systems konzipiert wurde. Die zentrale Herausforderung bestand darin, eine skalierbare und robuste Infrastruktur bereitzustellen, die sowohl **semantische Suche** als auch **kontextbasierte Antwortgenerierung** zuverlässig unterstützt.

#### Architekturprinzipien 

Die Backend-Architektur orientiert sich an folgenden Prinzipien:

* **API-First-Design**: Sämtliche Funktionalitäten werden über REST-APIs verfügbar gemacht, um eine klare Trennung zwischen Frontend und Backend sicherzustellen.
* **Asynchrone Verarbeitung**: Python-Async-Patterns gewährleisten eine performante Abwicklung von I/O-intensiven Aufgaben.
* **Modularität**: Komponenten wie Retrieval-Augmented Generation (RAG), Embeddings oder LLM-Integration sind als eigenständige Module implementiert.
* **Cloud-Native Deployment**: Containerisierung und Ausführung in der IBM Cloud sorgen für Skalierbarkeit und Wartbarkeit.

#### Datenfluss

Der Datenfluss folgt einem **Pipeline-Pattern**:

1. **Eingabe**: FastAPI empfängt und validiert Benutzeranfragen.
2. **Embeddings**: Texte werden mit dem Granite-Embedding-Modell in Vektoren umgewandelt.
3. **Semantische Suche**: Die Vektoren werden in PostgreSQL (pgvector) mit bestehenden Dokumenten abgeglichen.
4. **Kontextaggregation**: Relevante Dokumentenabschnitte werden zusammengestellt.
5. **Antwortgenerierung**: Das Granite-LLM erstellt Antworten unter Berücksichtigung des Kontexts.
6. **Quellenangabe**: Verwendete Dokumentenquellen werden transparent referenziert.

#### Technologie-Stack 

| Komponente        | Technologie                     | Zweck                                   |
| ----------------- | ------------------------------- | --------------------------------------- |
| **Web Framework** | FastAPI                         | REST-API und Request-Handling           |
| **Datenbank**     | PostgreSQL + pgvector           | Vektorspeicherung und semantische Suche |
| **Embeddings**    | IBM Granite Embedding 278M      | Text-zu-Vektor-Transformation           |
| **LLM**           | IBM Granite 3 8B Instruct       | Kontextbasierte Antwortgenerierung      |
| **Sprachdienste** | IBM Watson Speech-to-Text / TTS | Multimodale Eingabe und Ausgabe         |
| **Container**     | Docker                          | Deployment und Skalierung               |



### 4.2.2 FastAPI als Schnittstelle 

Das Backend wurde parallel zum Frontend entwickelt und fungiert als zentrale **Integrationsschicht**. **FastAPI** erwies sich als besonders geeignet, da es eine performante asynchrone Verarbeitung, automatische Dokumentation und eine klare Schnittstellendefinition ermöglicht.

Die Kernaufgaben des Backends umfassen:

* Entgegennahme und Verarbeitung von Benutzeranfragen
* semantisches Retrieval relevanter Dokumente
* kontextbasierte Antwortgenerierung
* transparente Quellenangaben

Als Datenbasis diente eine **PostgreSQL-Datenbank** mit **pgvector**, um semantische Suchen auf Embeddings zu ermöglichen. Damit entstand ein **RAG-System**, bei dem kontextrelevante Dokumententeile automatisch in die Antwortgenerierung einfließen.



### 4.2.3 Erste Hürden: Projektinstanz und Entwicklungsumgebung 

Der Einstieg in die Backend-Implementierung war von erheblichen Schwierigkeiten geprägt. Insbesondere das Anlegen einer Projektinstanz in der IBM Cloud führte wiederholt zu Problemen: Der Prozess blieb regelmäßig im Ladevorgang hängen und verhinderte so die Einrichtung einer stabilen Arbeitsumgebung. Dieses Problem ließ sich nur dadurch umgehen, dass ein Teammitglied die Instanz erfolgreich anlegte, wodurch jedoch eine **Abhängigkeit** entstand.

Die IBM-Cloud-Oberfläche wurde als wenig intuitiv wahrgenommen. Ohne Vorerfahrung war es herausfordernd, sich im komplexen Menüsystem zurechtzufinden. Viele Schritte mussten deshalb über das **Command Line Interface (CLI)** erfolgen, was zwar Flexibilität bot, aber die Fehlersuche erschwerte. Dokumentationen mussten intensiv herangezogen werden, was den Prozess erheblich verlangsamte.

Ein zusätzlicher Stolperstein war die Nutzung eines **geteilten Accounts**. Dies führte zu Verwechslungen mit einer anderen Gruppe und versehentlichem Arbeiten in deren Projektumgebung. Dabei wurden Teile fremder Projekte überschrieben, während auch im eigenen Projekt Dateien durch falsche Befehle beschädigt wurden. Die mangelnde Transparenz über den aktuellen Status sorgte für wiederholte Frustrationen.



### 4.2.4 Deployment, Logs und WatsonX-Integration 

Besonders komplex gestaltete sich die Einrichtung von **Build- und Deployment-Prozessen**. Anfangs war unklar, wie Fehler zu interpretieren waren, da die Log-Ausgaben nur schwer verständlich waren.

Ein zentrales Hindernis war die **Integration von WatsonX.ai**. Für die Nutzung der Modelle waren sowohl eine **Projekt-ID** als auch ein gültiger **API-Key** erforderlich. Wiederholt traten Fehlermeldungen auf, unter anderem:

* *„Json document validation error: project\_id should be a version 4 uuid“* – Projekt-ID im falschen Format.
* *„no\_associated\_service\_instance\_error“* – Projekt nicht mit Watson Machine Learning verknüpft.
* *„container\_not\_found“* – Projekt-ID nicht gefunden oder falsche Region.
* *„model\_not\_supported“* – Anfrage an ein nicht verfügbares Modell.

Auch beim Deployment der Container-Images in die IBM Container Registry kam es zu Problemen:

* *„invalid username/password: unauthorized“* – fehlende oder fehlerhafte Authentifizierung.
* *„denied: exceeded image storage quota“* – Speicherlimit der Registry war erreicht.

Diese Fehler führten zu wiederholtem Debugging und verzögerten die Entwicklung erheblich.

Trotz dieser Schwierigkeiten konnten schließlich zwei zentrale Modelle integriert werden:

* **Granite Embedding 278M Multilingual** für Vektorrepräsentationen
* **Granite 3 8B Instruct** für kontextbasierte Antwortgenerierung

Damit entstand ein funktionierendes RAG-System, das kontextspezifische Antworten mit Bezug zu den Onboarding-Dokumenten lieferte.



### 4.2.5 Positive Erfahrungen: Sprachdienste von IBM {#425-positive-erfahrungen-sprachdienste-von-ibm}

Neben den zahlreichen Hürden gab es auch **reibungslos funktionierende Bereiche**. Besonders positiv fiel die Anbindung der **Speech-to-Text-** und **Text-to-Speech-Dienste** von IBM auf. Die Integration über bereitgestellte Endpunkte verlief unkompliziert und erweiterte die Funktionalität des Systems über die rein textbasierte Interaktion hinaus. Dadurch wurden neue Nutzungsszenarien ermöglicht, etwa für barrierefreie Zugänge oder multimodale Schnittstellen.



Ergebnis: Das Backend von *Boardy* konnte trotz zahlreicher Herausforderungen erfolgreich umgesetzt werden. Die Integration moderner Technologien ermöglichte ein funktionierendes RAG-System, während die gesammelten Erfahrungen mit IBM Cloud und WatsonX wertvolle Erkenntnisse für zukünftige Entwicklungen liefern.








## 5. Fazit und kritische Reflexion 

Die Entwicklung der Boardy-Anwendung hat gezeigt, dass die grundlegende Zielsetzung – die Realisierung eines KI-gestützten Chat-Assistenten zur Unterstützung des Onboarding-Prozesses – technisch umsetzbar ist. Mit der Kombination aus **FastAPI**, **React**, **PostgreSQL mit pgvector** sowie **IBM WatsonX.ai** entstand ein funktionsfähiger Prototyp, der die in der Anforderungsanalyse definierten Kernfunktionen weitgehend erfüllt.

Die Evaluation ergab, dass das **RAG-System** relevante Dokumente zuverlässig abrufen konnte, auch wenn die Antworten nicht immer in einer optimal nutzerfreundlichen Form präsentiert wurden. Die durchschnittlichen **Antwortzeiten der API** erwiesen sich als praxistauglich, und die Implementierung der **multimodalen Eingabe** – bestehend aus Text, Sprachaufnahme und Dokument-Upload – gelang erfolgreich.

Trotz dieser Erfolge traten verschiedene Herausforderungen auf, die eine kritische Reflexion erforderlich machen. Die **Integration von WatsonX.ai** mit der Vektorsuche in PostgreSQL erwies sich als komplex und fehleranfällig, insbesondere im Hinblick auf asynchrone Programmierung, Projekt-IDs und die Verknüpfung mit Watson Machine Learning-Instanzen. Hinzu kam, dass die Wahl geeigneter Modelle (z. B. Granite vs. LLaMA) nicht immer eindeutig war und fehlerhafte Modell-IDs zu wiederholten Debugging-Schleifen führten. Der Einsatz von **KI-gestützten Code-Generatoren** wie Cursor oder Loveable AI beschleunigte zwar die Entwicklung, brachte jedoch zusätzlichen Aufwand durch unsauberen Code und erhöhte technische Schuld (vgl. Senn, 2019). Auch die Umsetzung des **Figma-Designs** in React machte deutlich, dass starre Designvorgaben aufwändig sein können, während iterative, toolgestützte Anpassungen oft zielführender sind.

Wichtige **Lerneffekte** konnten dennoch erzielt werden. Dazu zählen die Arbeit mit asynchroner Programmierung in Python, die Integration externer APIs sowie die Umsetzung multimodaler Schnittstellen. Besonders die Erfahrungen mit WatsonX.ai lieferten wertvolle Einsichten: Der Umgang mit Projekt-IDs, Service-Verknüpfungen und Modellverfügbarkeiten verdeutlichte, wie wichtig eine präzise Konfiguration und kontinuierliche Überprüfung ist. Zugleich wurde deutlich, dass bei der Nutzung generativer Werkzeuge eine Balance zwischen Entwicklungsgeschwindigkeit und langfristiger Codequalität gefunden werden muss.

Ein zentraler Aspekt des Projekts war die **Teamarbeit**. Die komplementäre Verteilung der Kompetenzen – UI-Design, Softwareentwicklung und WatsonX-Integration – ermöglichte eine effiziente Arbeitsteilung, erforderte aber intensive Abstimmung und gegenseitige Wissensvermittlung. Immer wieder entstanden Abhängigkeiten, die Zeitverluste nach sich zogen. Erschwerend kam hinzu, dass zu Beginn mehrere Teammitglieder nacheinander abwesend waren, wodurch sich die Entwicklung verzögerte.

Die **Projektorganisation** orientierte sich an agilen Methoden nach Scrum. Regelmäßige Stand-Ups, in der Endphase auf zwei Tage verdichtet, unterstützten die Koordination. Dennoch führten parallele Verpflichtungen und die fehlende formelle Projektleitung zu Verzögerungen und erschwerter Verantwortungszuweisung. Trotz dieser Umstände erwies sich die agile Arbeitsweise als hilfreich, um die komplementären Kompetenzen nutzbar zu machen und das Projekt erfolgreich abzuschließen.

Zusammenfassend lässt sich feststellen, dass die Boardy-Anwendung die **Machbarkeit eines KI-gestützten Onboarding-Assistenten** überzeugend demonstriert. Die Architektur ist stabil, die Kernfunktionen weitgehend umgesetzt, und die User Experience fällt positiv aus. Gleichzeitig wurde die Bedeutung klarer Kommunikation, iterativer Anpassungen und einer bewussten Balance zwischen Effizienz und Qualität deutlich.


## 5.1 Lessons Learned 

* **WatsonX.ai verstehen und korrekt integrieren**: Der Umgang mit Projekt-IDs, Service-Verknüpfungen und Modell-IDs stellte sich als Lernfeld heraus. Die Erfahrungen haben gezeigt, dass eine gründliche Dokumentation und Testumgebung notwendig sind, um Integrationsfehler frühzeitig zu erkennen.
* **Iteratives Vorgehen bevorzugen**: Starre Designs erschweren Anpassungen. Ein iteratives Vorgehen mit Testzyklen und Zwischenfeedback erhöht die Effizienz.
* **Technische Integration als Engpass**: Die Kombination von Vektorsuche, asynchroner Programmierung und externer Modell-Integration erwies sich als komplex. Klare Schnittstellendefinitionen sind für die Zukunft unverzichtbar.
* **Teamorganisation und Kommunikation**: Komplementäre Kompetenzen sind wertvoll, erfordern aber klare Rollen und kontinuierlichen Austausch. Frühzeitige Rollenklärung reduziert Abhängigkeiten und verhindert Zeitverluste.
* **KI-gestützte Entwicklung kritisch einsetzen**: Code-Generatoren sind hilfreich zur Beschleunigung, erfordern aber eine bewusste Nachbearbeitung, um langfristige Wartbarkeit zu gewährleisten.

## 5.2 Ausblick 

Für die nächste Entwicklungsphase ergeben sich konkrete **Weiterentwicklungspotenziale**:

* **Aktive Benachrichtigungen**: Boardy soll eigenständig Nachrichten an Mitarbeitende senden können, etwa Erinnerungen oder proaktive Hilfsangebote.
* **Nutzerprofile und Login**: Eine Personalisierung durch Profile ist notwendig, um Chats, Präferenzen und Fortschritte individuell speichern zu können.
* **Mehrere parallele Chats**: Die Strukturierung nach Themenbereichen erhöht die Übersichtlichkeit und Nutzbarkeit im Arbeitsalltag.
* **Verbesserte Nutzerfreundlichkeit**: Antworten müssen klarer, strukturierter und leichter verständlich werden. Prompt-Optimierung und Feedbackschleifen können dies unterstützen.
* **Informationsweitergabe an Führungskräfte**: Die Fähigkeit, offene Fragen oder neues Wissen automatisch aus Chats zu extrahieren und weiterzuleiten, soll implementiert werden.
* **Frontend-Optimierung**: Das aktuelle React-Frontend ist funktional, könnte aber durch gezielte Usability- und Design-Verbesserungen ein moderneres Nutzererlebnis bieten.


Ergebnis: Das Projekt hat die technische Machbarkeit und den Nutzen von Boardy erfolgreich unter Beweis gestellt, zugleich aber klare Entwicklungsfelder für die Zukunft offengelegt. Besonders die Erfahrungen mit WatsonX.ai liefern wertvolle Impulse, wie KI-gestützte Systeme robuster und anwenderfreundlicher integriert werden können.
