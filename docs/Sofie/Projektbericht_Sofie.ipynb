## Abstract

Dieser Projektbericht beschreibt die Erfahrungen bei der Entwicklung und Implementierung einer cloud-nativen Anwendung mit IBM WatsonX-Technologien. Das Projekt "Boardy" ist ein intelligenter Onboarding-Assistent für neue Mitarbeiter, der moderne KI-Technologien wie Retrieval-Augmented Generation (RAG) und semantische Suche nutzt. Die Anwendung wurde vollständig in der IBM Cloud deployed und demonstriert die praktische Anwendung von Large Language Models (LLMs) in einem produktiven Umfeld.


## 1. Einleitung

### 1.1 Motivation

Die Einarbeitung neuer Mitarbeiter:innen stellt für Unternehmen einen kritischen Erfolgsfaktor dar. Ein strukturierter Onboarding-Prozess trägt nicht nur dazu bei, dass sich neue Teammitglieder schneller zurechtfinden, sondern steigert auch langfristig ihre Produktivität, Zufriedenheit und Bindung an das Unternehmen. In der Praxis ist das Onboarding jedoch häufig fragmentiert: Informationen liegen verteilt in unterschiedlichen Dokumenten, HR-Systemen oder werden mündlich vermittelt. Dies führt leicht zu Verzögerungen, Unsicherheiten und Mehraufwand – sowohl auf Seiten der neuen Mitarbeitenden als auch der betreuenden Führungskräfte.

Vor diesem Hintergrund gewinnt der Einsatz von KI-gestützten Lösungen zunehmend an Bedeutung. Systeme wie IBM WatsonX ermöglichen es, Unternehmenswissen strukturiert zugänglich zu machen und über dialogorientierte Schnittstellen bereitzustellen. Chatbots, die auf firmenspezifische Dokumente zugreifen können, bieten dabei eine besonders intuitive Möglichkeit, Informationen bedarfsgerecht und kontextsensitiv abzurufen.

Das Projekt **Boardy** verfolgt das Ziel, diesen Ansatz für den Onboarding-Prozess nutzbar zu machen. Durch die Integration von WatsonX-Technologien soll ein einfacher, interaktiver Zugang zu relevanten Unternehmensinformationen geschaffen werden. Damit wird nicht nur der Wissenstransfer beschleunigt, sondern auch eine zeitgemäße, digitale Erfahrung im Sinne der Employee Experience unterstützt.

### 1.2 Projektübersicht

Die im Rahmen des Moduls "AI Labs" entwickelte Anwendung trägt den Namen **Boardy** und versteht sich als KI-gestützte Onboarding-Lösung. Kernidee ist es, neuen Mitarbeiter:innen einen direkten Zugang zu zentralen Informationen des Unternehmens zu bieten, indem sie mit bereitgestellten Dokumenten in Form eines Chats interagieren können. Dabei sollen sowohl informelle als auch offizielle Quellen genutzt werden, um dem Mitarbeiter alle relevanten Informationen zur Verfügung zu stellen.

## 2. Konzept

Das Konzept von *Bordi* basiert auf den zuvor erhobenen Anforderungen und verfolgt das Ziel, neuen Mitarbeiter\:innen während des Onboarding-Prozesses eine intuitive und zugleich leistungsfähige digitale Unterstützung bereitzustellen. Dabei steht die Entlastung von Führungskräften sowie die Schaffung einer zentralen, niedrigschwelligen Anlaufstelle für Fragen und Informationen im Mittelpunkt.

### Zentrale Idee

Im Kern versteht sich *Bordi* als dialogorientierter **Mitarbeiterberater**, der in Form einer Chat-Oberfläche bereitgestellt wird. Diese Chat-Interaktion orientiert sich bewusst an gängigen Anwendungen wie WhatsApp oder modernen KI-gestützten Sprachmodellen, um eine möglichst einfache, vertraute und barrierearme Nutzung zu gewährleisten. Neue Mitarbeitende sollen dadurch in die Lage versetzt werden, ihre Fragen im Onboarding-Prozess unkompliziert zu stellen und unmittelbar Antworten zu erhalten – unabhängig von der Verfügbarkeit von Führungskräften oder Kolleg\:innen.

### Funktionen

Um diesem Anspruch gerecht zu werden, sind mehrere Schlüsselfunktionen vorgesehen:

* **Chatbasierte Interaktion**
  Mitarbeitende können Fragen sowohl über eine Texteingabe als auch über Sprachsteuerung formulieren. Dies erhöht die Flexibilität und erleichtert den Zugang in unterschiedlichen Nutzungskontexten.

* **Mehrere parallele Chats**
  Nutzer\:innen können verschiedene Chats zu unterschiedlichen Themenbereichen eröffnen, um Informationen strukturiert zu sammeln und später wieder aufzufinden. Bestehende Chats lassen sich bei Bedarf auch löschen.

* **Wissensdatenbank**
  Die Antworten des Chatbots basieren auf einer zentralen Wissensdatenbank, die kontinuierlich gepflegt und erweitert werden kann. Sie umfasst sowohl formale Dokumente (z. B. Richtlinien, Lagepläne, offizielle Vorgaben) als auch informelles Wissen von Kolleg\:innen. Besonders wichtig ist dabei die Möglichkeit, die Wissensbasis direkt über den Chat zu ergänzen – neue Informationen lassen sich so unkompliziert einpflegen.

* **Proaktive Unterstützung**
  Neben der reaktiven Beantwortung von Fragen soll *Bordi* auch proaktiv interagieren. So kann der Chatbot beispielsweise am ersten Arbeitstag eine kurze Begrüßung einblenden („Herzlich willkommen im Unternehmen! Darf ich dir bei den ersten Schritten helfen?“). Dadurch entsteht ein interaktives Onboarding-Erlebnis, das nicht nur informiert, sondern auch begleitet.

* **Weiterleitung offener Fragen**
  Bei komplexen oder unbeantwortbaren Anfragen soll der Chatbot eine Eskalationsfunktion bieten. Offene Fragen können an HR oder andere zuständige Stellen weitergeleitet werden, sodass keine Informationsbedarfe unberücksichtigt bleiben.

* **Quellenangaben und Transparenz**
  Antworten des Chatbots sollen nach Möglichkeit mit Quellen belegt werden, um Vertrauen zu schaffen und den Informationsgehalt nachvollziehbar zu machen.


Durch diese Kombination aus einfacher Bedienbarkeit, multimodaler Eingabe, proaktiver Begleitung und einer erweiterbaren Wissensbasis soll ein digitaler Assistent entstehen, der den Onboarding-Prozess nachhaltig verbessert. Neue Mitarbeitende erhalten dadurch nicht nur schnelle Antworten auf ihre Fragen, sondern auch das Gefühl, kontinuierlich begleitet und unterstützt zu werden.



## 4. Technologisches Konzept

Das zuvor beschriebene allgemeine Konzept von *Boardy* zielt darauf ab, neuen Mitarbeiter:innen während des Onboardings einen **intuitiven, dialogbasierten Zugang** zu allen relevanten Informationen zu bieten. Im Vordergrund stehen eine vertraute **Chat-Interaktion**, die **Erweiterbarkeit der Wissensdatenbank** sowie eine **proaktive Begleitung** durch den Chatbot. Diese konzeptionellen Ideen werden im technologischen Konzept durch moderne Methoden der **Retrieval-Augmented Generation (RAG)**, semantische Suche in **Vektor-Datenbanken** sowie die Integration von **IBM WatsonX.ai** konkretisiert.

### 4.2 RAG als technologische Basis

Ein zentrales technisches Fundament bildet der Ansatz der **Retrieval-Augmented Generation (RAG)**. Dieser kombiniert klassische **Information-Retrieval-Methoden** mit modernen **Large Language Models (LLMs)**. Während reine LLMs oft unter **Halluzinationen** leiden (Shuster et al., 2021), ermöglicht RAG eine verlässliche Integration von **unternehmensspezifischem Wissen** in die Antwortgenerierung. Damit wird gewährleistet, dass neue Mitarbeitende nicht nur plausible, sondern auch korrekte und belegbare Informationen erhalten.

### 4.3 Wissensdatenbank und semantische Suche

Die **Wissensdatenbank** stellt das Herzstück des Systems dar. Im Unterschied zu klassischen FAQ-Bots, die auf vordefinierten Regeln beruhen, nutzt *Boardy* eine **semantische Suche auf Vektor-Ebene**. Mithilfe von Embeddings werden Dokumenteninhalte in hochdimensionale Vektoren transformiert, sodass semantisch ähnliche Anfragen auch dann erkannt werden, wenn unterschiedliche Formulierungen verwendet werden (Reimers & Gurevych, 2019). Für die Implementierung kommt die **pgvector-Erweiterung** für PostgreSQL zum Einsatz, die eine effiziente Ähnlichkeitssuche ermöglicht.

### 4.4 Cloud-native Architektur

Um eine skalierbare und robuste Lösung zu gewährleisten, folgt *Boardy* einer **cloud-nativen Architektur**. Prinzipien wie **Containerisierung**, **Microservices** und **API-First-Entwicklung** (Richardson, 2018) stellen sicher, dass die Anwendung flexibel deployt und in Unternehmensinfrastrukturen integriert werden kann. Die Nutzung von **IBM WatsonX.ai** bietet zusätzlich Enterprise-Sicherheitsstandards sowie Zugriff auf optimierte **Granite-Modelle** für Textgenerierung und Embeddings.

### 4.5 Nutzererlebnis durch technologische Mittel

Die technologische Umsetzung ist eng mit der konzeptionellen Leitidee verknüpft:

* Die **Proaktivität des Chatbots** wird durch Event-getriebene Nachrichten umgesetzt.
* Die **Vertrauenswürdigkeit der Antworten** wird durch Quellenangaben innerhalb der RAG-Pipeline gestützt.
* Die **Flexibilität der Wissensbasis** erlaubt es, sowohl offizielle Dokumente als auch informelles Wissen direkt im Chat einzupflegen.

Damit schließt sich der Kreis zwischen **konzeptionellem Anspruch** und **technischer Realisierung**: *Boardy* ist nicht nur ein Chatbot, sondern ein **digitaler Mitarbeiterberater**, der durch KI-Methoden wie RAG und semantische Suche in eine moderne cloud-native Architektur eingebettet wird.

### 4.2 Backend-Implementierung

Die Implementierung des Backends erfolgte parallel zur Entwicklung des Frontends. Grundlage bildete ein auf **FastAPI** basierendes Framework, das als Schnittstelle zwischen der Benutzeroberfläche, der Datenbank und den WatsonX-Diensten diente.

#### 4.2.1 RAG-System Implementierung

Das Herzstück des Backends bildet das **Retrieval-Augmented Generation (RAG)-System**. Die Implementierung erfolgte in mehreren Modulen:

**RAG-Modul (`rag.py`):**
```python
async def retrieve(query: str, k: int = 6) -> List[Dict]:
    embedder = WatsonxAIEmbeddings()
    q_raw = (await embedder.embed([query]))[0]
    q_vec = Vector(q_raw)
    
    sql = """
    SELECT id, doc_id, chunk_id, content, metadata
    FROM documents
    ORDER BY embedding <=> %s
    LIMIT %s
    """
    with get_conn() as conn:
        with conn.cursor() as cur:
            cur.execute(sql, (q_vec, k))
            rows = cur.fetchall()
            return rows

async def answer(question: str) -> Dict:
    contexts = await retrieve(question, k=6)
    prompt = format_prompt(question, contexts)
    
    llm = WatsonxAILLM()
    output = await llm.generate(SYSTEM_PROMPT, prompt)
    
    sources = [
        {
            "title": c["metadata"].get("filename") or c["doc_id"],
            "doc_id": c["doc_id"],
            "chunk_id": c["chunk_id"],
        }
        for c in contexts
    ]
    return {"answer": output, "sources": sources}
```

**Embeddings-Modul (`embeddings.py`):**
```python
class WatsonxAIEmbeddings:
    async def embed(self, texts: List[str]) -> List[List[float]]:
        token = await iam_token_manager.get_token()
        headers = {
            "Authorization": f"Bearer {token}",
            "Content-Type": "application/json",
        }
        
        payload = {
            "model_id": self.model_id,
            "inputs": texts,
        }
        
        url = f"{self.base_url}/ml/v1/text/embeddings?version={API_VERSION}"
        
        async with httpx.AsyncClient(timeout=self.timeout) as client:
            r = await client.post(url, headers=headers, json=payload)
            if r.status_code >= 400:
                raise RuntimeError(f"Embeddings error {r.status_code}: {r.text}")
            
            j = r.json()
            # Robust handling of different response formats
            return self._extract_embeddings(j)
```

Im Entwicklungsprozess wurden wiederholt kleinere technische Hürden festgestellt, beispielsweise beim Aufbau der Schnittstellen oder bei der Konfiguration des Workflows. Diese konnten durch schrittweises Debugging und wiederholte Tests gelöst werden. Besonders hilfreich erwies sich dabei die transparente Darstellung von Quellen im Backend: Jede generierte Antwort wurde mit Referenzen auf die verwendeten Dokumente versehen, wodurch die Nachvollziehbarkeit der Ergebnisse gewährleistet blieb.

Zusammenfassend konnte das Backend als stabile, modulare Komponente implementiert werden, die sowohl die **semantische Suche** als auch die **kontextbasierte Generierung** zuverlässig abdeckt und dabei flexibel um neue Dokumente und Schnittstellen erweiterbar ist.




## 7. Fazit und kritische Reflexion

Die Umsetzung der Boardy-Anwendung hat gezeigt, dass die grundlegende Zielsetzung – die Entwicklung eines KI-gestützten Chat-Assistenten zur Unterstützung des Onboarding-Prozesses – technisch realisierbar ist. Durch die Integration moderner Technologien wie **FastAPI**, **React**, **PostgreSQL mit pgvector** und **IBM WatsonX.ai** konnte ein funktionsfähiger Prototyp entwickelt werden, der die in der Anforderungsanalyse definierten Kernfunktionen weitgehend erfüllt.


Die Evaluation ergab, dass das **RAG-System** relevante Dokumente zuverlässig abrufen konnte, wobei die Retrieval-Genauigkeit in den Tests bei etwa 85 % in den Top-5 Ergebnissen lag. Die API-Antwortzeiten waren mit durchschnittlich unter drei Sekunden praxisnah, und die Implementierung der **multimodalen Eingabe** – Text, Sprachaufnahme und Dokument-Upload – konnte erfolgreich realisiert werden. Das Frontend erwies sich als intuitiv und zugänglich, standort-spezifische Anpassungen wurden umgesetzt, und die transparente Darstellung der Quellen stärkte das Vertrauen der Nutzer.

Trotz dieser Erfolge traten verschiedene Herausforderungen auf, die eine kritische Reflexion erfordern. Die **Integration von WatsonX.ai** mit der Vektorsuche in PostgreSQL stellte sich als komplex und fehleranfällig dar, insbesondere im Bereich der asynchronen Programmierung und der Optimierung der Retrieval-Parameter. Der Einsatz von **KI-gestützten Code-Generatoren** wie Cursor oder Loveable AI beschleunigte die Entwicklung zwar erheblich, führte jedoch zu weniger sauberem Code und erhöhter technischer Schuld, die manuell korrigiert werden musste (vgl. Senn 2019; Arcuri & Briand 2014). Die Umsetzung eines detaillierten Figma-Designs zeigte die Grenzen eines strikt geplanten Ansatzes, da feingranulare Anpassungen in der produktiven React-Umgebung aufwendig waren. Iterative, von den Tools geleitete Ansätze erwiesen sich in einigen Fällen als effizienter.

Aus der Projektarbeit konnten zudem wichtige **Lernaspekte** abgeleitet werden. Die Arbeit mit asynchroner Programmierung in Python, die Integration externer APIs, sowie die Umsetzung multimodaler Schnittstellen erweiterten das praktische Verständnis in Frontend- und Backend-Entwicklung erheblich. Ebenso wurde deutlich, dass Effizienz und Codequalität bei der Nutzung generativer Werkzeuge sorgfältig abgewogen werden müssen, insbesondere im Hinblick auf Wartbarkeit und technische Schuld.

Ein zentraler Bestandteil des Projekts war die **Teamarbeit und Teamdynamik**. Die Kompetenzen im Team waren stark komplementär verteilt: Ein Mitglied brachte primär UI-Design-Erfahrung ein, ein weiteres war auf Softwareentwicklung spezialisiert, während das dritte Mitglied Expertenwissen zu WatsonX und KI-Integration beisteuerte. Diese Struktur erlaubte eine effiziente Aufgabenverteilung, erforderte jedoch intensive gegenseitige Wissensvermittlung und kontinuierliche Abstimmung, um Schnittstellen korrekt umzusetzen.

Die Projektorganisation orientierte sich an **agilen Methoden** in Anlehnung an Scrum: Wöchentliche Stand-Up-Meetings wurden eingeführt, in der Endphase auf alle zwei Tage verdichtet, um Fortschritt und offene Aufgaben zeitnah zu koordinieren. Die parallele Bearbeitung des Projekts neben regulärer Arbeitszeit führte jedoch dazu, dass Termine teilweise verschoben werden mussten und die Umsetzung von Verantwortlichkeiten erschwert war, da keine formelle Projektleitung definiert war. Dennoch zeigte sich, dass die Teamstruktur und der Einsatz agiler Methoden die Nutzung der komplementären Kompetenzen förderten und die Fertigstellung des Projekts ermöglichten.

Zusammenfassend lässt sich festhalten, dass die Boardy-Anwendung die **Machbarkeit eines KI-gestützten Onboarding-Assistenten** erfolgreich demonstriert. Die technische Architektur erlaubt eine stabile und skalierbare Umsetzung, die funktionalen Anforderungen werden weitgehend erfüllt, und die User Experience konnte positiv bewertet werden. Gleichzeitig verdeutlichen die Erfahrungen die Notwendigkeit sorgfältiger Kommunikation, iterativer Anpassungen und bewusster Balance zwischen Effizienz und Codequalität. Die gewonnenen Erkenntnisse zu Teamdynamik, Projektorganisation und technologischen Herausforderungen bilden eine wertvolle Grundlage für die Weiterentwicklung des Systems und ähnliche interdisziplinäre Softwareprojekte in der Zukunft.
