## Abstract

Dieser Projektbericht beschreibt die Erfahrungen bei der Entwicklung und Implementierung einer cloud-nativen Anwendung mit IBM WatsonX-Technologien. Das Projekt "Boardy" ist ein intelligenter Onboarding-Assistent für neue Mitarbeiter, der moderne KI-Technologien wie Retrieval-Augmented Generation (RAG) und semantische Suche nutzt. Die Anwendung wurde vollständig in der IBM Cloud deployed und demonstriert die praktische Anwendung von Large Language Models (LLMs) in einem produktiven Umfeld.

## 2. Literaturrecherche und theoretische Grundlagen

### 2.1 Retrieval-Augmented Generation (RAG)

RAG ist ein Ansatz, der die Stärken von Information Retrieval und Textgenerierung kombiniert. Das System:

1. **Retrieval**: Sucht relevante Dokumente basierend auf der Benutzeranfrage
2. **Augmentation**: Erweitert den Kontext mit den gefundenen Informationen
3. **Generation**: Generiert eine Antwort unter Verwendung des erweiterten Kontexts

### 2.2 Vektor-Datenbanken und semantische Suche

Semantische Suche nutzt Embeddings, um die Bedeutung von Texten zu erfassen. Die pgvector-Erweiterung für PostgreSQL ermöglicht:

- Speicherung von hochdimensionalen Vektoren
- Effiziente Ähnlichkeitssuche mit COSINE-Distanz
- Skalierbare Vektor-Operationen

### 2.3 Cloud-native Architektur

Cloud-native Entwicklung folgt Prinzipien wie:

- **Containerisierung**: Docker für konsistente Deployment-Umgebungen
- **Microservices**: Lose gekoppelte, unabhängig deploybare Services
- **API-First**: RESTful APIs für Service-Kommunikation
- **DevOps**: Automatisierte CI/CD-Pipeline

### 2.4 IBM WatsonX.ai

IBM WatsonX.ai ist eine Enterprise-KI-Plattform, die verschiedene Large Language Models (LLMs) und Embeddings-Modelle bereitstellt:

- **Granite-Modelle**: IBM-eigene LLM-Familie
- **Multilingual Support**: Unterstützung für verschiedene Sprachen
- **Enterprise-Sicherheit**: Compliance und Datenschutz
- **API-Integration**: Einfache Integration in bestehende Anwendungen


## 3. Methodik und Systemarchitektur

### 3.1 System-Architektur

Das System folgt einer 3-Tier-Architektur mit klarer Trennung von Frontend, Backend und Datenbank:

```mermaid
graph TB
    A[Frontend - React] --> B[Backend - FastAPI]
    B --> C[PostgreSQL + pgvector]
    B --> D[WatsonX API]
    E[IBM Cloud Code Engine] --> B
    F[IBM Cloud PostgreSQL] --> C
    G[IBM WatsonX.ai] --> D
```

### 3.2 Technologie-Stack

| Komponente | Technologie | Version | Zweck |
|------------|-------------|----------|-------|
| Frontend | React + TypeScript | 18.2.0 + 5.4.0 | Benutzeroberfläche |
| Build Tool | Vite | 5.2.0 | Frontend-Build und -Entwicklung |
| Backend | FastAPI | 0.112.2 | REST-API und Business-Logik |
| Server | Uvicorn | 0.30.6 | ASGI-Server für FastAPI |
| Datenbank | PostgreSQL + pgvector | 3.1+ | Vektordatenbank für RAG |
| LLM | IBM Granite-3-8B | 8B Parameter | Textgenerierung |
| Embeddings | IBM Granite-Embedding | 278M Parameter | Semantische Suche |

### 3.3 RAG-Implementierung

Der RAG-Workflow umfasst folgende Schritte:

1. **Dokumentenverarbeitung**: Chunking und Embedding-Generierung
2. **Vektor-Speicherung**: Speicherung in PostgreSQL mit pgvector
3. **Query-Processing**: Embedding-Generierung für Benutzeranfragen
4. **Retrieval**: Semantische Suche nach relevanten Chunks
5. **Generation**: LLM-basierte Antwortgenerierung mit Kontext

Das Projekt folgte einem **iterativen Entwicklungsansatz** mit einer Scrumähnlichen Struktur:

1. **Anforderungsanalyse**: Gains, Pains, Jobs-Analyse

5. **Testing**: Kontinuierliche Tests und Optimierungen
6. **Deployment**: Cloud-native Bereitstellung

### 4.4 Deployment-Architektur

Das Deployment erfolgte über **IBM Cloud Code Engine**, einen Service, der containerisierte Anwendungen automatisch skaliert und verwaltet. Dadurch konnte ich mein gesamtes System – bestehend aus einem **FastAPI-Backend** und einem **React-Frontend** – in einem einzigen Multi-Stage-Docker-Image bereitstellen.

Der große Vorteil für mich war, dass ich mich nicht mehr mit einer eigenen Orchestrierung oder komplexer Infrastruktur beschäftigen musste. Der Service stellte mir direkt einen verschlüsselten **HTTPS-Endpunkt** zur Verfügung, über den sowohl das Backend als auch das Frontend unter derselben Domain erreichbar sind.

#### 4.4.1 Docker-Konfiguration

```dockerfile
# Multi-Stage Build für Frontend und Backend
FROM node:20-slim AS frontend-build
WORKDIR /app/frontend
COPY frontend/package.json frontend/package-lock.json ./
RUN npm ci --no-audit --no-fund
COPY frontend/ .
RUN npm run build

FROM python:3.11-slim AS backend
WORKDIR /app
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
COPY . .
COPY --from=frontend-build /app/frontend/build ./frontend/build
CMD ["uvicorn", "server:app", "--host", "0.0.0.0", "--port", "8000"]
```

## 6. Diskussion und Lessons Learned

### 6.1 Technische Herausforderungen

#### 6.1.1 WatsonX-Integration
Die größte Herausforderung war für mich die **Integration der WatsonX-API** mit dem RAG-System. Besonders die **Vektor-Suche** in PostgreSQL erforderte viel Experimentieren, bis ich die richtige Balance zwischen Performance und Genauigkeit gefunden hatte.

**Lösungsansätze:**
- Optimierung der Embedding-Dimensionen
- Fine-Tuning der Retrieval-Parameter
- Caching-Strategien für häufige Anfragen

#### 6.1.2 Asynchrone Programmierung
Ein weiterer wichtiger Lernpunkt war die **asynchrone Programmierung** in Python. Die Kombination aus FastAPI, async/await und den WatsonX-API-Calls erforderte ein tiefes Verständnis der asynchronen Programmierung.

**Erkenntnisse:**
- Proper error handling bei async operations
- Connection pooling für Datenbankverbindungen
- Timeout-Management für externe API-Calls

### 6.2 UI/UX-Entwicklung

#### 6.2.1 Design-Tool-Vergleich
Während der Entwicklung habe ich verschiedene Ansätze für das User Interface ausprobiert. Besonders interessant war der Vergleich zwischen verschiedenen Design-Tools:

**Manuelle React-Implementierung vs. AI-Tools:**
- **React**: Vollständige Kontrolle, bessere WatsonX-Integration
- **Loveable AI**: Schnelle Prototypen, aber limitierte Funktionalität
- **Fazit**: Manuelle Implementierung für komplexe KI-Integration überlegen

#### 6.2.2 Multimodale Eingabe
Die Implementierung der **multimodalen Eingabe** war technisch herausfordernd, aber sehr wertvoll für die Benutzerfreundlichkeit:

- **MediaRecorder API**: Browser-Kompatibilität beachten
- **File Upload**: Sicherheitsaspekte und Validierung
- **Responsive Design**: Mobile-First Ansatz

### 6.3 Performance-Optimierungen

Ich habe verschiedene Optimierungen implementiert:

- **Bundle-Splitting** im Frontend für schnellere Ladezeiten
- **Connection Pooling** für die Datenbank
- **Token-Caching** für WatsonX-API-Calls
- **Lazy Loading** für große Dokumente

### 6.4 Sicherheit und Compliance

**Implementierte Sicherheitsmaßnahmen:**
- **API-Keys**: Verwaltung über Environment Variables
- **Datenbank**: SSL-verschlüsselte Verbindung
- **Logging**: Maskierung sensibler Daten
- **CORS**: Konfiguriert für Frontend-Domains

So konnte ich sicherstellen, dass vertrauliche Informationen geschützt bleiben.

### A.1 Code-Beispiele

#### A.1.1 RAG-Implementierung

```python
async def answer(question: str) -> Dict:
    contexts = await retrieve(question, k=6)
    prompt = format_prompt(question, contexts)
    
    llm = WatsonxAILLM()
    output = await llm.generate(SYSTEM_PROMPT, prompt)
    
    sources = [
        {
            "title": c["metadata"].get("filename") or c["doc_id"],
            "doc_id": c["doc_id"],
            "chunk_id": c["chunk_id"],
        }
        for c in contexts
    ]
    return {"answer": output, "sources": sources}
```

#### A.1.2 Frontend-Integration

```typescript
const handleSendMessage = (text: string) => {
  const apiBaseUrl = import.meta.env.VITE_API_BASE_URL;
  fetch(`${apiBaseUrl}/v1/ask`, {
    method: 'POST',
    headers: { 'Content-Type': 'application/json' },
    body: JSON.stringify({ query: text })
  })
    .then((r) => r.json())
    .then((data) => {
      setMessages(prev => [...prev, {
        id: crypto.randomUUID(),
        text: data.answer,
        isUser: false,
        timestamp: new Date(),
        sources: data.sources || []
      }]);
    });
};
```

### A.2 Deployment-Konfiguration

#### A.2.1 Dockerfile

```dockerfile
FROM node:20-slim AS frontend-build
WORKDIR /app/frontend
COPY frontend/package.json frontend/package-lock.json ./
RUN npm ci --no-audit --no-fund
COPY frontend/ .
RUN npm run build

FROM python:3.11-slim AS backend
WORKDIR /app
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
COPY . .
COPY --from=frontend-build /app/frontend/build ./frontend/build
CMD ["uvicorn", "server:app", "--host", "0.0.0.0", "--port", "8000"]
```
